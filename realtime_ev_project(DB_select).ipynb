```python
import time

def thread_run():
    threading.Timer(900, thread_run).start()
    print('**id_start**')
    print('====',time.ctime(),'====')
    sid=[]
    #충전소_딕셔너리
    c_station = {'sid':[],'name':[], 'u_time':[], 'agency': [],'do':[], 'city':[], 'gu':[], 'c_add':[],'phone':[], 'fee':[],'etc':[], 'create_date':[]}
    #충전기_딕셔너리
    charger = {'cid':[],'c_name':[], 'typ':[], 'charge':[], 'c_using':[], 'amount':[], 'c_create_date':[], 'sid':[]}
        
    for k in range(2,19):
        driver.find_element_by_css_selector('#F_SI_DO1')
        driver.implicitly_wait(5)
        driver.find_element_by_css_selector('#F_SI_DO1 > option:nth-child(%s)'%k).click()
        driver.find_element_by_css_selector('#statList')
        driver.implicitly_wait(30)
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')

        a = soup.find_all('div',class_='fw_path')
        for link in a:
            temp = str(link['sid'])
            sid.append(temp)
    global tmp_len
    global tmp_len_2
    global cd
    global c_cd
    global count
    
    c_station['create_date']=cd
    
    if tmp_len != len(sid):
        tmp_len2 = len(sid)-tmp_len
        for i in range (0,tmp_len2):
            i = time.ctime()
            c_station['create_date'].append(i)
        

    tmp_len = len(sid)
    
    
    print('**id_finish**')
    print('====',time.ctime(),'====')
    
    print('**date_start**')
    print('====',time.ctime(),'====')
    for i in sid:
        url_base = 'https://www.ev.or.kr/portal/monitor/evminfo?sid='
        url = url_base+str(i)
        html = urlopen(url)
        soup = BeautifulSoup(html, 'html.parser')

        for tmp_using in soup.find_all(class_='table_01'):
            n_tmp = soup.find_all('h2')
            c_station['name'].append(n_tmp[0].get_text())
            c_station['sid'].append(i)

            t_tmp = soup.find_all('em')
            c_station['u_time'].append(t_tmp[0].get_text())

            info_tmp = soup.find_all(class_='table_02')
            dinfo = info_tmp[0].find_all('td')
            c_station['c_add'].append(dinfo[0].get_text())
            c_station['agency'].append(dinfo[1].get_text())
            c_station['phone'].append(dinfo[2].get_text())

            tmp_fee = dinfo[3].get_text()
            tmp_fee = re.sub('&nbsp; | &nbsp; |\n|\t|\r|', '', tmp_fee)
            c_station['fee'].append(tmp_fee)


            tmp_etc = dinfo[5].get_text()
            tmp_etc = re.sub('&nbsp; | &nbsp; |\n|\t|\r|', '', tmp_etc)
            c_station['etc'].append(tmp_etc)


            do_tmp = dinfo[0].get_text().split(' ')
            do_tmp = do_tmp[0]
            c_station['do'].append(do_tmp)
            gu_tmp = dinfo[0].get_text().split(' ')
            if '도' in do_tmp:
                city_tmp = dinfo[0].get_text().split(' ')[1]
                if '군' in dinfo[0].get_text().split(' ')[1]:
                    gu_tmp = gu_tmp[1]
                else:
                    gu_tmp = gu_tmp[2]
                c_station['gu'].append(gu_tmp)
                c_station['city'].append(city_tmp)
            elif '특별'or'광역' in do_tmp:
                do_tmp = do_tmp[0]+do_tmp[1]
                gu_tmp = gu_tmp[1]
                c_station['gu'].append(gu_tmp)
                c_station['city'].append(do_tmp+'시')

            ev_tmp = tmp_using.find(class_='ev_type t03')
            if len(tmp_using)<=5:

                n_tmp = soup.find_all('h2')
                charger['c_name'].append(n_tmp[0].get_text())
                charger['cid'].append(i+'0')
                charger['sid'].append(i)

                ty_tmp = soup.find_all('dd')
                charger['typ'].append(ty_tmp[0].get_text())

                tmp_amount = dinfo[4].get_text()

                tmp_amount = re.findall('\d+', tmp_amount)
                if tmp_amount==[]:
                    charger['amount'].append(0)
                else:
                    charger['amount'].append(tmp_amount[0])

                for u_tmp in soup.find_all(class_='td3'):
                    tmp = re.sub('\n', '', u_tmp.get_text())
                    charger['c_using'].append(tmp)


                td2 = soup.find_all(class_='td2')
                for te in td2[0:]:
                    test=[]
                    for a in te.find_all('span'):
                        tmp=[]
                        if len(a.get_text())<=6:
                            tmp.append(a.get_text())
                            test+=tmp
                    test= str("+".join(test))
                    charger['charge'].append(test)

            elif ev_tmp == None:
                td2 = soup.find_all(class_='td2')
                t=0
                for te in td2[0:]:
                    test=[]
                    for a in te.find_all('span'):
                        tmp=[]
                        tmp.append(a.get_text())
                        test+=tmp
                    test= str("+".join(test))
                    charger['charge'].append(test)

                    n_tmp = soup.find_all('h2')
                    charger['c_name'].append(n_tmp[0].get_text())
                    charger['cid'].append(i+str(t))
                    t+=1
                    charger['sid'].append(i)
                    ty_tmp = soup.find_all('dd')
                    charger['typ'].append(ty_tmp[0].get_text())

                    tmp_amount = dinfo[4].get_text()

                    tmp_amount = re.findall('\d+', tmp_amount)
                    if tmp_amount==[]:
                        charger['amount'].append(0)
                    else:
                        charger['amount'].append(tmp_amount[0])

                    for u_tmp in soup.find_all(class_='td3'):
                        tmp = re.sub('\n', '', u_tmp.get_text())
                        charger['c_using'].append(tmp)
                        break        

            elif len(tmp_using)>6:
                tmp_using = soup.find_all(class_='ev_type t03')
                t=0
                for te in tmp_using[0:]:
                    for a in te.find_all('span'):
                        if a.get_text()==' ':
                            break
                        elif len(a.get_text())<=6:
                            charger['charge'].append(a.get_text())

                            n_tmp = soup.find_all('h2')
                            charger['c_name'].append(n_tmp[0].get_text())
                            charger['cid'].append(i+str(t))
                            t+=1
                            charger['sid'].append(i)
                            ty_tmp = soup.find_all('dd')
                            charger['typ'].append(ty_tmp[0].get_text())

                            tmp_amount = dinfo[4].get_text()

                            tmp_amount = re.findall('\d+', tmp_amount)
                            if tmp_amount==[]:
                                charger['amount'].append(0)
                            else:
                                charger['amount'].append(tmp_amount[0])


                            for u_tmp in soup.find_all(class_='td3'):
                                tmp = re.sub('\n', '', u_tmp.get_text())
                                charger['c_using'].append(tmp)
                                break
                                
    charger['c_create_date']=c_cd
    
    if tmp_len_2 != len(charger['c_name']):
        tmp_len3 = len(charger['c_name'])-tmp_len_2
        for i in range (0,tmp_len3):
            i = time.ctime()
            charger['c_create_date'].append(i)

    tmp_len_2 = len(charger['c_name'])
    
    print('**data_finish**')
    print('====',time.ctime(),'====')
    print('충전소개수:',len(c_station['name']))
    print('충전기개수:',len(charger['c_name']))
    print('사용현황:',len(charger['c_using']))
    print('create_date:',len(c_station['create_date']))
    print('c_create_date:',len(charger['c_create_date']))
    
    
    #db저장
    print('**DB_start**')
    print('====',time.ctime(),'====')
    
    tmp_station={'sid':[],'name':[], 'u_time':[], 'agency': [],'do':[], 'city':[], 'gu':[], 'c_add':[],'phone':[], 'fee':[],'etc':[], 'create_date':[]}
    #tmp_charger={'cid':[],'c_name':[], 'typ':[], 'charge':[], 'c_using':[], 'amount':[], 'sid':[], 'c_cd:[]'}
    
    conn = pymysql.connect(host='127.0.0.1', user='root', password='1234', db='ev_project', charset='utf8')
    cur = conn.cursor()
    
    sql="SELECT * FROM charging_station"
    
    st =[]
    cur.execute(sql)
    result = cur.fetchall()
    for i in result:
        tmp_station['sid'].append(i[0])
        tmp_station['name'].append(i[1])
        tmp_station['u_time'].append(i[2])
        tmp_station['agency'].append(i[3])
        tmp_station['do'].append(i[4])
        tmp_station['city'].append(i[5])
        tmp_station['gu'].append(i[6])
        tmp_station['c_add'].append(i[7])
        tmp_station['phone'].append(i[8])
        tmp_station['fee'].append(i[9])
        tmp_station['etc'].append(i[10])
        tmp_station['create_date'].append(i[11])
        
    sql="select * from charger"
    df = pd.read_sql(sql, conn)

    tmp_charger={'cid':df['cid'],'c_name':df['c_name'], 'typ':df['typ'], 'charge':df['charge'], 'c_using':df['c_using'], 'amount':df['amount'], 'c_create_date':df['c_create_date'], 'sid':df['sid']}
    
    print('DB충전소개수:',len(tmp_station['name']))
    
    #충전소
    if len(tmp_station['name']) != 0:
        for t in tmp_station.keys():
            for j in range(0, len(tmp_station['name'])):
                if tmp_station[t][j]!=c_station[t][j]:
                    sql=("update charging_station set "+t+"='%s' where "+t+"='%s' and sid='%s'") %(c_station[t][j], tmp_station[t][j], c_station['sid'][j])
                    cur.execute(sql)
                #if cd[j]!=create_date[j]:
                #    sql="update charging_station set create_date='%s' where name='%s'"%(create_date[j], tmp_station['name'][j])
                #    cur.execute(sql)
    
    if len(tmp_station['name'])!=len(c_station['name']):
        for i in range(0,len(c_station['name'])):
            sql="""insert into charging_station select '%s','%s', '%s', '%s','%s','%s','%s','%s','%s','%s','%s','%s' from dual where not exists (select * from charging_station where sid='%s')""" % (c_station['sid'][i], c_station['name'][i], c_station['u_time'][i], c_station['agency'][i],c_station['do'][i], c_station['city'][i], c_station['gu'][i], c_station['c_add'][i], c_station['phone'][i], c_station['fee'][i], c_station['etc'][i], c_station['create_date'][i], c_station['sid'][i])
            cur.execute(sql)

    #충전기
    
    global status
    
    date=time.strftime('%Y-%m-%d ')+time.ctime().split(' ')[0]
    u_time=time.strftime("%H:%M:%S")
    
    
    if len(tmp_charger['c_name']) != 0:
        for t in tmp_charger.keys():
            for j in range(0, len(tmp_charger['c_name'])):
                if tmp_charger[t][j]!=charger[t][j]:
                    sql=("update charger set "+t+"='%s' where "+t+"='%s' and cid = '%s'")%(charger[t][j], tmp_charger[t][j], charger['cid'][j])
                    cur.execute(sql)
    
    sql="select * from usage_status"
    df1 = pd.read_sql(sql, conn)
    
    tmp_status={'id':df1['id'],'start_date':df1['start_date'], 'start_time':df1['start_time'], 'end_date':df1['end_date'], 'end_time':df1['end_time']}

    for i in range(0,len(charger['cid'])):
        for j in range(0,len(tmp_status['id'])):
            if tmp_status['id'][j]==tmp_charger['cid'][i] and tmp_status['end_date'][i] == None:
                if tmp_charger['c_using'][i]=='사용가능':
                    sql="update usage_status set end_date='%s', end_time='%s' where id='%s'"%(date, u_time, charger['cid'][i])
                    cur.execute(sql)
            elif tmp_status['id'][j]==tmp_charger['cid'][i] and tmp_status['end_date'][i] != None:
                if tmp_charger['c_using'][i]=='사용중':
                    sql="""insert into usage_status (id, start_date, start_time) values ('%s', '%s', '%s') """ % (charger['cid'][j], date ,u_time)
                    cur.execute(sql)
        
    if len(tmp_charger['c_name'])!=len(charger['c_name']):
        for i in range(0,len(charger['c_name'])):
            sql="""insert into charger (cid, sid, c_name, typ, charge, c_using, amount, c_create_date) select '%s','%s','%s', '%s', '%s','%s','%s','%s' from dual where not exists (select * from charger where cid='%s')""" % (charger['cid'][i],charger['sid'][i],charger['c_name'][i], charger['typ'][i], charger['charge'][i],charger['c_using'][i],charger['amount'][i], charger['c_create_date'][i], charger['cid'][i])
            cur.execute(sql)
            if count==0:
                if charger['c_using'][i]=='사용중':
                    status.append('S')
                    sql1="""insert into usage_status (id, start_date, start_time) values ('%s', '%s', '%s') """ % (charger['cid'][i], date ,u_time)
                    cur.execute(sql1)

        count+=1
    cd = c_station['create_date']
    c_cd = charger['c_create_date']
    conn.commit()
    conn.close()
    
    print('====',time.ctime(),'====')
    print('**DB finish**')
    print('사용현황(status):',len(status))
    
    threading.daemon=True
    #threading.Timer(20, thread_run).start()
    
thread_run()
```
